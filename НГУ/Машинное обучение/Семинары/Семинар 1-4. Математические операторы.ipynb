## Новые операции

* **Целочисленное деление**, записывается как два знака деления подряд "//" , как следует из названия цель такого деления - отбросить дробную часть результата.


```python
print(5 // 2)
```

```python
print(11.9999999 // 3)
```

* В противоположность прошлой операции существует, **остаток от деления**,  записывается знаком процента и возвращает остаток. 


```python
print(18 % 3)
```

Важно заметить операция возвращает не дробную часть результата, а именно остаток изначально числа. 


image.png

Допустим мы хотим узнать продолжительность фильма который идет 196 минут 

```python
a = 196

print("Фильм идет", a // 60, "часа и", a % 60, "минут" )
```
