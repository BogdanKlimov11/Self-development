## Цикл for


Синтаксис следующий:

```
for i in диапазон:
    тело цикла
```
Для получения диапазона чисел часто используется функция `range()`.


Код, который будет выполняться каждую итерацию цикла, записывается после двоеточия с отступом 4 пробела от строчки c `for`.

Посмотрим на примере:

````python
x = 0

for i in range(365):
  x = x + 1
  print("Нам нужно сделать", x,"подтягиваний")
````

````python
for i in range(1,101,2):
  print(i)
````

## Break

В Python оператор break позволяет прервать цикл при возникновении внешнего фактора.


````python
print("Угадай число которое я загадал.")
print("У тебя всего 5 попыток.")

for i in range(5):
  a = int(input("Введи свое число: "))

  if a > 15:
    print("Меньше")
  elif a == 15:
    print("Угадал")
    break
  else:
    print("Больше")
````

````python
print("Угадай число которое я загадал.")
print("У тебя всего 5 попыток.")

for i in range(5):
  a = int(input("Введи свое число: "))

  if a > 15:
    print("Меньше")
  elif a == 15:
    print("Угадал")
    print("Поздравляю ты выйграл")
    break
  else:
    print("Больше")

else:
  print("Попытки кончились")
  print("Ты проиграл")
````

Оператор continue позволяет пропустить часть цикла при возникновении внешнего фактора и перейти к следующей итерации цикла.

````python
string = 'Мир. Труд. Май.'

for letter in string:
  if letter == ".":
    continue

  print(letter)
````
