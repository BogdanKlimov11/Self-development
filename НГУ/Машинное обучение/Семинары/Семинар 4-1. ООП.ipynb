## Cоздадим свой первый обьект


```python
class Lamp:
    status=False
    brightness=5
```

**Класс** это тип данных, который определен в языке, а **объект** — переменная этого класса.


```python
Lamp_hallway=Lamp()
```

```python
print("Статус лампочки", Lamp_hallway.status)
print("Ee яркость", Lamp_hallway.brightness)
```

```python
class Lamp:
  status=False
  brightness=5

  def info(self):  # обрати внимание на self

      print("Статус лампочки", self.status)
      print("Eе яркость", self.brightness)
```

```python
Lamp_hallway=Lamp()

Lamp_hallway.info()
```

### Что такое **self** ?

**Self** используют для представления объекта класса. Благодаря нему объект может получить доступ к атрибутам и методам своего класса.


Получается, когда мы пишем подобную строчку:





```python
Lamp_hallway.info()
```

Для объекта она выглядит так:


```python
Lamp.info(Lamp_hallway)
```

### Атрибуты и поля

Функция написана внутри класса и справляется со своей задачей. Вот только если попытаться вызвать её не при помощи объекта, а напрямую, программа выдаст ошибку:


```python
info()
```

Чтобы не возникало путаницы, все **функции**, помещённые внутрь объектов, называют **методами**.  Давай посмотрим их отличия:

Переименование не обошло и переменные. Смысл переменных —  хранить значение так, чтобы его можно было посмотреть в любой момент. При этом переменная внутри объекта принадлежит только ему. **Переменные** внутри объекта называют **параметром** или **полем** класса.
