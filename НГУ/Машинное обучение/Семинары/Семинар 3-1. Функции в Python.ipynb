```python
def sum_squares(a,b):
  c = a**2 + 2*b*a + b**2
  return c
```

```python
result = sum_squares(5,6)

print(result)
```

```python
def avg(x):
  summ = 0
  for i in x:
    summ=summ+i

  return summ/len(x)
```

```python
a=[25,56,37,98,200,105]
 
print(avg(a))
```
