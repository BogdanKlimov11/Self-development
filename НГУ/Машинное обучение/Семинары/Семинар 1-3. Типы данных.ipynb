## Типы данных

image.png

Рассмотрим 3 базовых типа 

* Целочисленный тип данных, в статьях его называют int сокращение от английского слова (**integer**)   
* Дробный тип данных или число с плавающей запятой **float**;
* Строка или набор символов **string**.

В python переменная автоматически принимает тот тип данных который в нее подается. Т.е если написать:



```python
a = "Привет!"
```

то переменная **а**, автоматически станет переменной типа string

```python
pi = 3.1415
```

то переменная **pi**, закономерно получит тип float.

## Операции над строками

* Сложение

```python
word1 = "Я помню "
word2 = "чудное многвенье"

print(word1 + word2)
```

* Умножение на число

```python
word1 = "Текст"
number1 = 9

print(word1 * number1)
```

Любая другая операция вызовет ошибку

```python
word1 = "Я помню "
word2 = "чудное многвенье"

print(word1 - word2)
```

## Явная типизация


```python
variable1 = input("Введите первое слагаемое ")
variable2 = input("Введите второе слагаемое ")

print(variable1 + variable2)
```

В этом нам помогут специальные функции, по одной на каждый тип данных
* str() — возвращает аргумент, переданный в скобках, в виде строки (текста);
* int() — возвращает аргумент, переданный в скобках, в виде целого числа;
* float() — возвращает аргумент, переданный в скобках, в виде десятичной дроби.

А если ты не уверен какой тип данных у переменной используй функцию:
* type() — возвращает тип аргумента, указанного в скобках;


```python
print(type(variable1))
```

```python
variable1 = input("Введите первое слагаемое ")
variable1 = int(variable1)

variable2 = input("Введите второе слагаемое ")
variable2 = int(variable2)


print(variable1 + variable2)
```

Такую программу можно и нужно **сократить**

```python
variable1 = int(input("Введите первое слагаемое "))
variable2 = int(input("Введите второе слагаемое "))

print(variable1 + variable2)
```

**Важно!**


Не всякую строку можно перевести в число. Если вы попытаетесь перевести в число строку, которая содержит буквы, это приведет к ошибке.


```python
name = int(input("Введи свое имя "))

print("Привет", name)
```
