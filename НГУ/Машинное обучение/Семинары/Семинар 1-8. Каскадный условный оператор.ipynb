## Каскадный условный оператор (плохой пример)

В приложении одного популярного сервиса по доставке еды, действовала акцию на бесплатную доставку при заказе от 2 000 рублей. Это легко отобразить на схеме:


image.png

```python
price_amount = 3000

if price_amount > 2000:
  delivery_count = 0
else:
  delivery_count = 200
```

```python
print(delivery_count)
```

С притоком новых пользователей акция изменилась и стала ступенчатой теперь при заказе:
* до 1000 рублей, доставка составит 200 рублей
* до 2000 рублей, доставка составит 100 рублей
* от 2000 рублей доставка будет бесплатной.


Новая схема алгоритма теперь выглядит так:


image.png

```python
price_amount = int(input())

if price_amount > 2000:
  delivery_count = 0
else:
  if price_amount > 1000:
     delivery_count = 100
  else:
    delivery_count = 200
```

```python
print("Цена доставки", delivery_count, "рублей")
```

## Вложенсть оператора

Может выглядеть так:

image.png

Или так:


image.png

```python
a = int(input())
b = int(input())
operation = input("Введите символ операции ")

if operation == "+":
  print(a+b)
else:
  if operation == "-":
    print(a-b)
  else:
    if operation == "*":
      print(a*b)
    else:
      if operation == "/":
        if b == 0:
          print("На ноль делить нельзя!")
        else:
          print(a/b)
      else:
        print("Неверная операция")
```

## Каскадный условный оператор (хороший пример)


image.png

Каскадный условный упрощает работу когда в Python нужно проверить несколько условий. Код прошлой задачи превращается  компактную запись вида:


```python
a = int(input())
b = int(input())
operation = input()

if operation == "+":
  print(a+b)
elif operation == "-":
  print(a-b)
elif operation == "*":
  print(a*b)
elif operation == "/":
  if b == 0:
    print("Ha ноль делить нельзя!")
  else:
    print(a/b)
else:
  print("Heверная операция!")
```

Давай по подробнее расмотрим структуру оператора


image.png

Т.е. если в такой конструкции одно из условий сработало, то все остальные отбрасываются, что экономит время работы у программы. На основе этого правила можно сделать два вывода:
В первые строчки оператора имеет смысл ставить наиболее популярные события (если конечно это возможно)
Каскадным условным оператором лучше проверять значения одной переменной или пары значений
