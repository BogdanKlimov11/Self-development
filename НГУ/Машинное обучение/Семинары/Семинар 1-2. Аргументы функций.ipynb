## Функция input()

Ну а теперь давай знакомится с главным напарником print(), функцией input()!

` input() — функция чтения данных с клавиатуры.`


Ввод осуществляется с помощью клавиатуры, окончание ввода обозначается нажатием клавиши “Enter”.


```python
input("Введи свое имя: ")
```

Поместим функцию **input**() внутрь **print**().


```python
print("Привет", input("Введи свое имя: "))
```

## Переменные

Переменная -  некая "коробка" в которую можно записать одно значение и в нужный момент считать его.


image.png

Чтобы создать переменную ей нужно дать имя. Хорошим тоном будет создание переменных c “говорящим” названием. Создадим переменную с именем number  и сохраним в ней число 9.


```python
number = 9
```

Запись значения в переменную выглядит следующим образом :   
> **имя_переменной** *= значение или выражение, результат которого мы хотим хранить в переменной.*


```python
print(number)
```

```python
name = input("Введи свое имя: ")
second_name = input("Введи свою фамилию: ")
```

```python
print("Привет", name, second_name, "не плохой день для программирования!")
```

Пременные как элементы уравнений

```python
a = 10
b = 37
```

```python
print(a+b)
```

При создании переменной стоит помнить о правилах оформления имен. Они:
* могут содержать только английские буквы и цифры
* не должны начинаться с цифры;
* не должны повторять уже встроенные в Python имена, например, print, str, int и т.д.;.


```python
a = "Как "
b = "дела"
c = "?"

print(a+b+c)
```

**Важно!**


Python чувствителен к регистру букв при написании кода. Это значит, что переменные Text, text и TEXT — абсолютно разные для интерпретатора кода.
