## Деструктор объекта

__del__() — деструктор объектов, который вызывается для их удаления. Помимо этого после окончания программы Python освобождает память и удаляет все объекты автоматически:


```python
class Simple_Class():
  def __del__(self):
    print("Мой час пришел")
```

```python
new_class = Simple_Class()
```

Для "ручного" удаления обьекта используется команда **del**

```python
del new_class
```

## Оператор сложения

__add__() — метод перегрузки оператора сложения. Когда объект будет выступать в качестве слагаемого с левой стороны, класс запустит код, который находится внутри оператора:


```python
class Simple_Class():
  def __add__(self,n):
    print("Думали ко мне прибавить",n,"?")
    print("А вот и нет")
```

```python
new_class = Simple_Class()
```

```python
b = new_class + 5
```
