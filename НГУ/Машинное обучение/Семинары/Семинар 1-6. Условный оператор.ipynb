В Python проверка условия осуществляется при помощи оператора if. Рассмотрим код для проверки пароля:


````python
password = input("Введите свой пароль: ")

if password == "qwerty123":
  print("Пароль правильный добро пожаловать")
````

Зафиксируем структуру ветвления:


```
if логическое условие:
    		результат, если условие выполняется
```







## Логическое ветвление в Си

Отступы - обязательный синтаксический инструмент при работе на языке Python.  Если в Си для указания принадлежности блока кода к условию он помешается фигурные кавычки. 


image.png

В Python в качестве этого выступают отступы.

````python
age = int(input("Сколько вам лет?: "))

if age < 18:
  print("Вы родились после 2003 года")
  print("Веротяно вы еще учитесь в школе")
  print("Зато вы молоды душой и духом!")
````

## Полный условный оператор


image.png

````python
password = input("Введите свой пароль: ")

if password == "qwerty123":
  print("Пароль правильный добро пожаловать")
else:
  print("Пароль неправиьлный, попробуй еще раз!")
````

В новой программе обрабатываются все случаи : 
* когда пользователь ввел правильный пароль, выражение возвращает True  и на экран выводится сообщение “Пароль правильный”,

* когда пароль не совпадает, выражение возвращает False, на экран выводится сообщение “Пароль неправильный”,
